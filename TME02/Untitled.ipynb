```python
def getRSV_bool(query,list_id,index_inverse):
    tf_query=normalizer.getTextRepresentation(query)
    pertinants=list_id
    for stem in tf_query.keys():
        pertinants= list(filter(lambda x : x in index_inverse[stem].keys(),pertinants))
    rsv_dic={id_doc : 1 if id_doc in pertinants else 0 for id_doc in list_id}
    return rsv_dic
    
```

```python
def getRSV_vect(query,list_id,index_inverse):
    tf_query=normalizer.getTextRepresentation(query)
    rsv_dic={doc:0 for doc in list_id}
    for stem in tf_query.keys():
        for id_doc in index_inverse[stem].keys():
            rsv_dic[id_doc]+=tf_query[stem]*index_inverse[stem][id_doc]
    return rsv_dic
```
